```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as ml
from mpl_toolkits.mplot3d import Axes3D
import os
print("libraries imported !")
```

```python
f = open("C:\\Users\\Shrishti D Hore\\OneDrive\\Documents\\Stir_Research\\corrosion_data_new.csv")
data = np.genfromtxt("C:\\Users\\Shrishti D Hore\\OneDrive\\Documents\\Stir_Research\\corrosion_data_new.csv", skip_header = 1, delimiter =',')
print("dataset for Corrosion pi potential/ mv imported !!")
```

```python
for line in f:
    print(line)
```

```python
#defining co-ordinates
x,y  = np.meshgrid(data[:,0], data[:,1])
z = np.tile(data[:,2], (len(data[:,2]), 1))
```

```python
#setting plot as 3d projection and defining plot as contour
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.contour(x, y,z, cmap = plt.get_cmap('rainbow'))
plt.colorbar()
plt.title("Corrosion Resistance Contour")
plt.show()
```
